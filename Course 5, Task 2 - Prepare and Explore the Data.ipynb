```python
import pandas as pd
```

```python
credit = pd.read_csv('default of credit card clients.csv', header = 1)
```

```python
credit.head()
```

```python
credit.describe()
```

```python
credit.info()
```

```python
#Remove the column 'ID' from the dataframe
credit.drop(columns = 'ID')
```

```python
#Check for missing Values
credit.isnull().values.any()
```

```python
#Change to nominal/categorical data type
credit["SEX"] = credit["SEX"].astype('category')
credit["EDUCATION"] = credit["EDUCATION"].astype('category')
credit["MARRIAGE"] = credit["MARRIAGE"].astype('category')
credit[['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']] = credit[['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']].astype('category')
credit['default payment next month'] = credit['default payment next month'].astype('category')
```

```python
credit.info()
```

```python
#Discretize age, group into 10 bins
pd.cut(credit['AGE'], 8)
```

```python
#Save updated dataframe as a csv
credit.to_csv('default updated.csv')
```
